### Splitting time series data - to avoid leakage

1. Train Set would be from 2012-19 - Clean training 
2. Dev set would be from 2020-21 - For hyperparameter tuning
3. Test set would be 2022-23 - Unbiased testing data

```python
df_train = df.loc[df.date.dt.year < 2020]
df_dev = df.loc[(df.date.dt.year >= 2020) & (df.date.dt.year <= 2021)]
df_test = df.loc[(df.date.dt.year >= 2022) & (df.date.dt.year <= 2023)]

print("Train shape : ", df_train.shape)
print("Dev shape : ", df_dev.shape)
print("Test shape : ", df_test.shape)
```
